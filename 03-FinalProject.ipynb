## Abstract

This project examines whether SAT Math and SAT Verbal median scores differ across U.S. regions and between public and private universities in 2023, and how these differences are related to admissions outcomes. Understanding these patterns is important because standardized test scores continue to play a role in college admissions, even in the context of test-optional policies. Prior research suggests that SAT performance is shaped not only by individual academic ability but also by regional inequalities in educational resources and differences in institutional selectivity. As a result, comparing SAT score distributions across regions and institution types can provide insight into how structural and institutional factors are reflected in admissions-related metrics.

To investigate this question, we used a dataset containing university-level information on median SAT Math and Verbal scores, institutional characteristics (including region and public vs. private status), and admissions outcomes such as acceptance rates. We conducted exploratory data analysis to compare score distributions across groups and to examine patterns of variation between Math and Verbal components. In addition, we analyzed how these score differences are associated with admissions outcomes by comparing acceptance rates across categories. This approach allows us to examine how academic metrics and institutional factors are related within the broader admissions landscape, without assuming a single underlying cause for observed differences.

Using university level data on SAT Math and Verbal medians, region, institution type, and admit rate, the analysis utilized exploratory visualizations with correlation, linear regression, and two way ANOVA to compare scores across groups and investigate whether those differences were statistically significant. 

The results suggest that both SAT Math and Verbal medians vary significantly by region and by public versus private, private schools generally reporting higher scores than public schools in most regions, especially in the west and northeast, while the midwest had smaller differences. SAT scores were also strongly associated with admit rate, so schools with higher median SAT scores tended to be more selective, though this relationship was slightly stronger for Math than Verbal scores. This analysis suggested that rising SAT math scores were associated with a larger drop in admit rates for private schools versus for public ones, meaning test score differences could be tied not only to regional variation in academic profiles but also to school selectivity. Our findings supported the conclusion that both geography and institution type are related to SAT distributions and admissions outcomes in 2023.

## Background and Prior Work

Admission decisions in U.S. higher education are shaped by the interaction of academic metrics, institutional selectivity, and broader structural inequalities. Among these factors, standardized test scores—particularly SAT Math and Evidence-Based Reading and Writing (Verbal)—have long served as key indicators of college readiness. Although many universities adopted test-optional policies by 2023, SAT scores remain widely reported and continue to influence both applicant strategies and institutional decision-making. As a result, examining variation in SAT Math and Verbal median scores across U.S. regions and between public and private universities provides a useful lens for understanding differences in academic preparation as well as patterns of admissions selectivity. Institutional datasets, such as those compiled by the University of California system, demonstrate how admissions outcomes (e.g., acceptance rates) can be analyzed alongside applicant academic characteristics, enabling integrated examination of both inputs and outcomes in the admissions process.<a href="#ref1">1</a>

A central insight from prior research is that SAT scores are not purely measures of individual academic ability but are strongly shaped by contextual factors. Recent large-scale analyses show that school-level and regional characteristics—including funding disparities, demographic composition, and access to college preparatory resources—significantly influence average SAT performance across geographic areas.<a href="#ref4">4</a> These findings suggest that observed regional differences in SAT Math and Verbal scores may reflect unequal access to educational opportunities rather than intrinsic differences in student ability, which is important for interpreting score variation within a broader structural context. Complementary evidence from University of California–focused research further indicates that socioeconomic background variables, such as family income and parental education, account for a substantial portion of variation in SAT scores and have become increasingly correlated with test performance over time.<a href="#ref5">5</a> Together, these studies highlight the importance of considering underlying structural factors when comparing SAT medians across regions.

Institutional context introduces an additional dimension to the interpretation of SAT score distributions. Public and private universities differ in their missions, applicant pools, and enrollment priorities, which may influence both observed score distributions and admissions outcomes. Research on highly selective private institutions shows that geographic and informational barriers can limit applications from high-achieving, low-income students, even when their SAT scores are competitive.<a href="#ref6">6</a> This suggests that differences in SAT medians across institution types may arise not only from selectivity but also from variation in applicant pool composition and recruitment practices. Considering these factors is important for understanding how institutional structures shape both who applies and how academic metrics are represented among admitted students.

At the same time, the role of SAT scores in admissions decisions has evolved in recent years. Policy-oriented analyses emphasize a shift toward holistic and contextualized evaluation, in which test scores are considered alongside GPA, coursework rigor, and indicators of socioeconomic background.<a href="#ref2">2</a> However, empirical evidence indicates that SAT scores continue to function as meaningful signals in the admissions process. For example, recent findings from the National Bureau of Economic Research show that submitting standardized test scores can increase admission probabilities, particularly for students from disadvantaged backgrounds.<a href="#ref7">7</a> This suggests that even in test-optional contexts, SAT scores may still play a role in shaping admissions outcomes, making it relevant to examine how score distributions relate to measures such as acceptance rates.

In addition to their role in admissions, standardized test scores are also associated with post-enrollment academic performance. Research from Opportunity Insights finds that SAT and ACT scores are predictive of college grades, although their predictive strength varies across institutions and student populations.<a href="#ref3">3</a> These findings indicate that SAT scores capture certain dimensions of academic preparation, while also reflecting broader contextual influences. Moreover, differences between SAT Math and Verbal components may represent distinct aspects of student preparation, which could vary across regions or institutional contexts due to differences in curriculum emphasis or educational resources.

Taken together, prior work suggests that both regional context and institutional characteristics shape observed SAT score distributions, but potentially through different mechanisms. Regional variation is often associated with differences in educational resources and opportunity structures, while institutional differences may reflect selection processes and applicant pool composition. This distinction is important for framing comparisons of SAT medians across regions and institution types, as it highlights multiple factors that may contribute to observed patterns.

Despite extensive research on standardized testing, several gaps remain. First, much of the literature focuses on composite SAT scores rather than explicitly comparing Math and Verbal components, limiting understanding of subject-specific variation. Second, prior studies often examine score distributions or admissions outcomes separately, rather than considering how they relate to one another. Third, while regional disparities and institutional differences are well documented independently, their interaction in shaping both SAT medians and admissions outcomes remains less explored in recent data.

Building on this literature, this project investigates whether SAT Math and SAT Verbal median scores differ significantly across U.S. regions and between public and private universities in 2023, and how these differences relate to admissions outcomes such as acceptance rates. By examining these factors jointly, the analysis aims to better understand how academic metrics, regional context, and institutional characteristics are associated within the broader admissions landscape.

## Hypothesis


We hypothesize that SAT Math and SAT Verbal median scores differ significantly across U.S. regions and between public and private universities in 2023. Specifically, we expect private universities to report higher median SAT scores than public universities, and we also expect measurable regional differences in both SAT components. ### University Admissions and SAT Outcomes by Region and Institution Type (IPEDS 2023 Combined Dataset)

This project combines two raw IPEDS 2023 files because neither file alone is sufficient for our research question. The admissions file (`adm2023_RV.csv`) contains SAT and admissions outcome variables but does not include the full contextual labels needed for grouped comparisons. The institutional header/characteristics file (`hd2023.csv`) provides those context variables, including institution name, state, region code, and control type. We merge these two sources by `UNITID` (IPEDS institution ID), so each final row still represents one institution but now includes both outcomes and grouping variables required for analysis.

The key outcome metrics are `SATMT50` and `SATVR50`, which are median SAT Math and SAT Verbal/Reading section scores at each institution. These are measured in SAT section points, where valid section values are generally between 200 and 800. We also retain `SATMT25`, `SATMT75`, `SATVR25`, and `SATVR75` to represent score spread within institutions. For admissions context, we keep `APPLCN` (number of applicants), `ADMSSN` (number admitted), and `ENRLT` (number enrolled), which are count variables. We also use `SATPCT` (0–100), the percentage of enrolled first-time students who submitted SAT scores, to evaluate how representative SAT statistics are at each institution.

The grouping variables come from institutional characteristics. `CONTROL` identifies institution control (public, private nonprofit, private for-profit), and `OBEREG` is the IPEDS region code. For this project, `OBEREG` is mapped to four broader U.S. regions (Northeast, Midwest, South, West). We also collapse institution control to `Public` vs `Private`. The reason is both conceptual and practical: after filtering out schools without usable SAT medians, one private subgroup (especially private for-profit) has very few institutions left, which makes separate subgroup comparisons unstable and noisy. Combining nonprofit and for-profit into a single private category gives more reliable group sizes for statistical comparison.

There are important limitations and potential biases. First, this is school-level aggregated data rather than student-level data, so analyses identify institutional associations rather than individual-level causal effects. Second, SAT fields have substantial missingness, likely related to test-optional and reporting practices, so missingness is probably systematic rather than random; this can bias the final sample. Third, `SATVR50` represents verbal/reading performance, not a direct writing score, so conclusions should use official variable definitions. Finally, collapsing private institution types improves stability but can hide differences between private nonprofit and private for-profit schools.

```python
import pandas as pd
import numpy as np

# Load raw files
adm = pd.read_csv("data/00-raw/adm2023_RV.csv", low_memory=False)
hd = pd.read_csv("data/00-raw/hd2023.csv", low_memory=False)

# Merge admissions with institution metadata
df_withoutadm = adm.merge(
    hd[["UNITID", "INSTNM", "STABBR", "OBEREG", "CONTROL"]],
    on="UNITID",
    how="left"
)

print("Raw shape:", df_withoutadm.shape)
print("\nMissingness (%):")
print((df_withoutadm.isna().mean() * 100).sort_values(ascending=False).round(2))

# Keep columns used for analysis
df_withoutadm = df_withoutadm[[
    "UNITID", "INSTNM", "STABBR", "OBEREG", "CONTROL",
    "APPLCN", "ADMSSN", "ENRLT", "SATPCT",
    "SATVR50", "SATMT50", "SATVR25", "SATVR75", "SATMT25", "SATMT75"
]].copy()

# Convert numeric columns
num_cols = ["APPLCN", "ADMSSN", "ENRLT", "SATPCT", "SATVR50", "SATMT50", "SATVR25", "SATVR75", "SATMT25", "SATMT75"]
for c in num_cols:
    df_withoutadm[c] = pd.to_numeric(df_withoutadm[c], errors="coerce")

# Collapse control into Public vs Private only
df_withoutadm["control_group"] = df_withoutadm["CONTROL"].map({1: "Public", 2: "Private", 3: "Private"})

# Build 4-region labels from OBEREG
region_map = {
    1: "Northeast", 2: "Northeast",
    3: "Midwest", 4: "Midwest",
    5: "South", 6: "South",
    7: "West", 8: "West",
    9: "Outlying"
}
df_withoutadm["region"] = df_withoutadm["OBEREG"].map(region_map)

# Drop schools without SAT medians and keep valid groups
df_withoutadm_clean = df_withoutadm.dropna(subset=["SATMT50", "SATVR50"]).copy()
df_withoutadm_clean = df_withoutadm_clean[df_withoutadm_clean["control_group"].isin(["Public", "Private"])].copy()
df_withoutadm_clean = df_withoutadm_clean[df_withoutadm_clean["region"].isin(["Northeast", "Midwest", "South", "West"])].copy()

# Keep SAT values in valid section-score range
df_withoutadm_clean = df_withoutadm_clean[
    df_withoutadm_clean["SATMT50"].between(200, 800) &
    df_withoutadm_clean["SATVR50"].between(200, 800)
].copy()

# Optional context metrics
df_withoutadm_clean["admit_rate"] = np.where(df_withoutadm_clean["APPLCN"] > 0, df_withoutadm_clean["ADMSSN"] / df_withoutadm_clean["APPLCN"], np.nan)
df_withoutadm_clean["yield_rate"] = np.where(df_withoutadm_clean["ADMSSN"] > 0, df_withoutadm_clean["ENRLT"] / df_withoutadm_clean["ADMSSN"], np.nan)

# Save cleaned data
df_withoutadm_final = df_withoutadm_clean.drop(columns=["CONTROL", "OBEREG"], errors="ignore")
df_withoutadm_final.to_csv("data/02-processed/university_sat_region_control_2023.csv", index=False)


print(df_withoutadm_clean.shape)
df_withoutadm_final.head()
```

### Relationship between SAT scores and Admissions Rate (Part 1) 
These scatterplots examine the relationship between institutional SAT medians (`SATMT50`, `SATVR50`) and admission rate, with points colored by region or institution type. We include this section because our updated question asks not only whether SAT medians differ by region/type, but also whether those SAT patterns are related to admissions outcomes.

Across all four plots, the dominant pattern is a negative association: institutions with higher SAT medians tend to have lower admission rates. This is visible for both Math and Verbal, and the pattern remains when points are colored by either region or control group. In practical terms, schools with higher reported SAT medians are generally more selective. Region and institution type appear to shift the concentration of points, but they do not remove the overall downward trend.

The scatterplots also show substantial spread in the mid-SAT range (roughly 500–650), where admission rates vary widely across schools. That spread suggests SAT medians are related to admissions outcomes but are not the only factor. This is expected because admission rates are influenced by multiple institutional and applicant-market factors.

```python
# MATH REGION 
sns.scatterplot(
    data=df_withoutadm,
    x='SATMT50',
    y='admit_rate',
    hue='region',  
)
plt.title('SAT math median vs admission rate(region)(figure 1)')
plt.xlabel('Median SAT Math score')
plt.ylabel('Admission Rate')
plt.legend()
plt.show()
# MATH INSTITUTION_TYPE
sns.scatterplot(
    data=df_withoutadm,
    x='SATMT50',
    y='admit_rate',
    hue='control_group',  
)
plt.title('SAT math median vs admission rate(Institution type)(figure 2)')
plt.xlabel('Median SAT Math score')
plt.ylabel('Admission Rate')
plt.legend()
plt.show()

# VERBAL REGION 
sns.scatterplot(
    data=df_withoutadm,
    x='SATVR50',
    y='admit_rate',
    hue='region',  
)
plt.title('SAT Verbal median vs admission rate(region)(Figure 3)')
plt.xlabel('Median SAT Verbal score')
plt.ylabel('Admission Rate')
plt.legend()
plt.show()

# VERBAL INSTITUTION_TYPE
sns.scatterplot(
    data=df_withoutadm,
    x='SATVR50',
    y='admit_rate',
    hue='control_group',  
)
plt.title('SAT Verbal median vs admission rate(Institution type)(figure 4)')
plt.xlabel('Median SAT Verbal score')
plt.ylabel('Admission Rate')
plt.legend()
plt.show()

from scipy.stats import pearsonr
math_r, math_p = pearsonr(df_withoutadm['SATMT50'], df_withoutadm['admit_rate'])
verbal_r, verbal_p = pearsonr(df_withoutadm['SATVR50'], df_withoutadm['admit_rate'])
print("SAT Math: ", math_r, "p-value:", math_p)
print("SAT Verbal: ", verbal_r, "p-value:", verbal_p)
```

## Ethics

Our project uses publicly available, institutional data on 2023 SAT Math and Verbal medians, US regions, institution types (public and private), and admissions outcomes. Since the dataset is aggregated at the university level, it doesn’t contain any personally identifiable information or individual student records. Due to this, there are little issues about direct privacy exposure. However, there are still ethical concerns in how data can be interpreted & presented. 

One of these concerns involves overinterpreting aggregate patterns. Differences in median SAT scores across regions and between public and private universities  do not justify making conclusions about individual students, or making conclusions about certain students in a certain region/group being more “capable” or “deserving” than other students. Our analysis involves institutional patterns, not individual capabilities, so we avoid language/wording that could encourage stereotyping or misleading interpretations. 

Another concern is biases in data representation. In 2023, many colleges were still test optional, so reported SAT medians could reflect the smaller group of applicants that submitted scores rather than reflecting the entire applicant pool. Additionally, differences across regions and institution types are shaped by a number of factors, including school resources, access to test preparation resources, family income, recruitment/application strategies, applicant self selection, and admissions policies. Due to this, our findings should be interpreted more as descriptive associations rather than causal claims. 

We also aim to represent data honestly and responsibly. We intend to clearly label variables, distinguish rates and counts, and avoid misleading representations that exaggerate differences in our visualizations and summaries. To avoid misinterpretation, we documented data sources, cleaning steps, and showed our analysis process so our project could be reproducible and so any issues/concerns could be easily traced and corrected. Since our project will be publicly shared on GitHub, we made sure to make our limitations explicit. Our data is aggregated, cross sectional, and shaped by test optional reporting, and isn’t appropriate for making individual level conclusions. If any figure or interpretation seems misleading or encourages harmful generalizations, we would revise wording, add any clarification, or remove misleading/problematic content.
