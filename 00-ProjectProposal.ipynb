## Research Question

What is the demographic distribution of first-generation and international undergraduate students admitted to UC San Diego (2018-2023), and how do these populations intersect across ethnicity, gender, and intended major categories?



## Background and Prior Work

Understanding patterns of college admissions and representation is essential for evaluating equity in public higher education. Within the University of California (UC) system, admissions policies play a major role in shaping student demographics and access to academic opportunities. UC San Diego (UCSD), as one of the most selective public universities in the United States, serves as an important case study for examining representation among groups such as first-generation and international students. Public UC admissions datasets provide information on enrollment trends by residency status, ethnicity, gender, and academic interest, allowing researchers to analyze how institutional policies and broader social factors influence student representation.</a><a href="#ref2">2</a>

A major historical influence on UC admissions policies is Proposition 209, passed in 1996, which prohibited the consideration of race, ethnicity, or sex in public university admissions. This policy significantly altered admissions strategies across the UC system and remains central to discussions about educational equity. Research conducted by the University of California Office of the President indicates that universities adopted alternative diversity strategies after Proposition 209, including expanded outreach efforts, holistic application review, and increased emphasis on socioeconomic factors. These changes demonstrate the importance of examining how demographic representation has evolved following major policy reforms.<a href="#ref1">1

Empirical research has explored the long-term effects of Proposition 209 and access-oriented admissions policies on underrepresented students. Studies suggest that while overall access to UC campuses has expanded, demographic representation across campuses and academic disciplines remains uneven, particularly among historically underrepresented minority groups.<a href="#ref3">3</a> Additionally, recent reporting indicates increasing enrollment of California residents alongside growth in international student populations, reflecting shifting institutional priorities and financial considerations.<a href="#ref4">4</a> Educational policy discussions also highlight ongoing barriers faced by California students, including disparities in academic preparation, access to counseling resources, and financial constraints that disproportionately affect first-generation applicants.

Building on prior research, this project examines the demographic distribution of first-generation and international undergraduate students admitted to UC San Diego between 2018 and 2023. By analyzing how these populations intersect across ethnicity, gender, and intended major categories, this study aims to provide a clearer understanding of how admissions policies and institutional trends influence student diversity at the campus level. Focusing on UCSD allows for a more detailed evaluation of representation across academic disciplines and contributes to broader discussions about accessibility and equity in public higher education.

## Hypothesis




We hypothesize that acceptance and yield rates will differ from international, out-of-state, and in-state student populations at UCSD from 2020 to 2025. Specifically, due to costs and travel restrictions, we believe international students will have a lower percentage yield than that of in-state and out-of-state students over the 5 year span. We also expect that acceptance rates will differ across these groups over these years in accordance with differences in finances and application priorities.


We expect that International students and out-of-state students will have lower yield percentages because of the higher tuition costs of going to UCSD relative to in-state counterparts. Additionally international students have student visas and travel constraints that may force accepted students to not be able to attend. This uncertainty in the admissions process will develop lower yield results over the 5 year span between each residency group. Additionally, We expect that admissions rates across groups will differ significantly due to application priorities and admissions pools. As a public university UC San Diego must prioritize in-state admissions, but when looking at application priorities, International and out-of-state students may have advantageous academic profiles due to the diversity of their programs and an ability to pay the higher tuition costs further funding the school.
